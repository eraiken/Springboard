## 2 - Features and their definitions <a name="features"></a>

Before diving further into our analysis, let's define the features we are analyzing.

### 2.1 Features from Spotify data <a name="featspot"></a>
Our Spotify dataframe, `songs_data_filtered_df`, includes the following variables, which we have defined by referring to the [Spotify Web API Reference](https://developer.spotify.com/documentation/web-api/reference/get-audio-features):
- `'popularity'`: A measure of a song's popularity
- `'is_explicit'`: Boolean variable indicating `0` for "Not explicit" and `1` for "Is explicit"
- `'duration_ms'`: The length of a song in milliseconds
- `'danceability'`: A continuous variable measured from 0.0 to 1.0, where 0.0 is least danceable and 1.0 is most danceable.
- `'energy'`: Represents "a perceptual measure of intensity and activity." Measured from 0.0 to 1.0, where 0.0 represents a low energy (slow and quiet) song and 1.0 represents a high energy ("fast, loud, and noisy") song.
- `'loudness'`: The overall loudness of a track in decibels (dB). Generally, the quietest songs are -60 dB while the loudest are 0 dB.
- `'mode'`: Boolean variable indicating the modality of a track, where `0` represents the minor scale and `1` represents the major scale.
- `'speechiness'`: A measure from 0 to 1 detecting "speech-like" recordings. According to the Spotify API documentation, values "above 0.66 describe tracks that are probably made entirely of spoken words. Values between 0.33 and 0.66 describe tracks that may contain both music and speech, either in sections or layered, including such cases as rap music. Values below 0.33 most likely represent music and other non-speech-like tracks."
- `'acousticness'`: According to Spotify API documentation, "A confidence measure from 0.0 to 1.0 of whether the track is acoustic. 1.0 represents high confidence the track is acoustic."
- `'instrumentalness'`: A measure from 0.0 to 1.0 predicting whether a track contains no vocals. "The closer the instrumentalness value is to 1.0, the greater likelihood the track contains no vocal content."
- `'liveness'`: A measure from 0.0 to 1.0 detecting the presence of an audience in the recording. "A value above 0.8 provides strong likelihood that the track is live."
- `'valence'`: A measure from 0.0 to 1.0 "describing the musical positiveness conveyed by a track. Tracks with high valence sound more positive" while "tracks with low valence sound more negative."
- `'tempo'`: "The overall estimated tempo of a track in beats per minute (BPM)."

Our dataframe also includes features `'spotify_id'`, `'name'`, `'album_name'`, `'artists'`, `'album_release_date'`, and `'key'`, and `time_signature`. These features might be helpful in answering interesting questions later on in our analysis (for example, among the happiest countries, who is the most popular artist? Who is the least popular artist?), but we will be ignoring these at this stage in our data analysis.

Something to note is that `valence` is an explicit measure of "positiveness" &mdash; i.e., how "happy" a track is. It will be interesting to see whether a meaningful correlation exists between valence and measures of world happiness.

### 2.2 Features from World Happiness data <a name="featwh"></a>
Our World Happiness dataframe, `wh_data`, uses data from the Gallup World Poll (GWP). (See [World Happiness Report 2024](https://www.kaggle.com/datasets/ajaypalsinghlo/world-happiness-report-2024) on Kaggle.) To define these features, we will be referencing the statistical appendix of the 2024 World Happiness Report (WHR), which can be found on the official [World Happiness Report website](https://happiness-report.s3.amazonaws.com/2024/Ch2+Appendix.pdf). This dataframe includes the following features:
- `'ladder_score'`: The "ladder score" is the "national average response to the question of life evaluations." This is how each survey respondent answered the question, "Please imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you. On which step of the ladder would you say you personally feel you stand at this time?"
- `'dystopia_residual'`: The baseline score for happiness in a hypothetical country called Dystopia, which has values equal to the world's lowest national averages for each of the six factors (listed below).
- `'eb_log_gdp'`, `'eb_social_support'`, `'eb_life_expectancy'`, `'eb_life_choice_freedom'`, `'eb_generosity'`, `'eb_corruption'`: The `eb_` portion in each of these variable names stands for "Explained by." These columns describe the extent to which these factors contribute in evaluating the happiness in each country. The factors include the following:
    - **'Logged GDP per capita'**: The log of the gross domestic product per capita.
    - **'Social support'**: The national average of binary responses (0 or 1) to the GWP question, "If you were in trouble, do you have relatives or friends you can count on to help you whenever you need them, or not?"
    - 'Healthy life expectancy': Measured in years, based on data from the World Health Organization's Global Health Observatory data repository.
    - **'Freedom to make life choices'**: The national average of responses to the GWP question, "Are you satisfied or dissatisfied with your freedom to chose what you do with your life?"
    - **'Generosity'**: Per the 2024 WHR report, "\[g\]enerosity is the residual of regressing national average of response to the GWP question, 'Have you donated money to a charity in the past month?' on GDP per capita."
    - **'Perceptions of corruption'**: The national average of survey responses to two questions from the GWP: "Is corruption widespread throughout the government or not?" and "Is corruption widespread within businesses or not?" The answers to these questions are binary (0 or 1) responses, and the overall perception is the average of these two responses.

The sum of these contributions, plus the Dystopia residual, equals the total happiness score.

## 3 - Initial EDA <a name="initial-eda"></a>

### Analysis of happiness components <a name="wh-happiness-components"></a>

As explained in the previous section, the sum of all "Explained by" factors, plus the Dystopia residual, equals the happiness score. Let's examine how each of these six factors contribute to the average composition of our countries' happiness scores.

```python
# Create a list of happiness components to use in our analysis
happiness_components = ['eb_log_gdp', 'eb_social_support', 'eb_life_expectancy',
                       'eb_life_choice_freedom', 'eb_generosity', 'eb_corruption']
```

```python
# Understanding the happiness score composition
wh_data['total_explained'] = wh_data[happiness_components].sum(axis=1) + wh_data['dystopia_residual']

# Verification check
print("Verification of Happiness Score Components:")
print(f"Average absolute difference between ladder score and total explained: "
      f"{(wh_data['ladder_score'] - wh_data['total_explained']).abs().mean():.6f}")

# Calculate percentage contribution of each component
for component in happiness_components + ['dystopia_residual']:
    wh_data[f'{component}_pct'] = (wh_data[component] / wh_data['ladder_score']) * 100

# Show average percentage contributions
pct_columns = [col for col in wh_data.columns if col.endswith('_pct')]
print("\nAverage Percentage Contribution to Happiness Score:")
print(wh_data[pct_columns].mean().round(2))
```

### Analysis of musical features vs. happiness <a name="musical-features-vs-happiness"></a>

```python
# Base features (weighted averages) we'll analyze
base_features = ['popularity', 'is_explicit_pct', 'mode_pct', 'duration_ms', 'danceability', 'energy', 'loudness', 
                 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
```

```python
# Standard deviation features
std_features = [col for col in songs_data_filtered_stats_country.columns if col.endswith('_std')]
```

```python
# Diversity features
diversity_features = [col for col in songs_data_filtered_stats_country.columns if col.endswith('_diversity')]
```

```python
def analyze_correlations(analysis_df, features, feature_type):
    """Analyze correlations between a set of features and happiness score"""
    correlations = {}
    p_values = {}
    
    for feature in features:
        corr, p_val = stats.pearsonr(analysis_df[feature], analysis_df['ladder_score'])
        correlations[feature] = corr
        p_values[feature] = p_val
    
    results_df = pd.DataFrame({
        'correlation': correlations,
        'p_value': p_values
    }).round(3)
    
    # Sort by absolute correlation strength
    results_df = results_df.reindex(results_df['correlation'].abs().sort_values(ascending=False).index)
    
    # Add significance indicators
    results_df['significance'] = ['***' if p < 0.001 else '**' if p < 0.01 
                                else '*' if p < 0.05 else '' for p in results_df['p_value']]
    
    print(f"\n{feature_type} Correlations with Happiness:")
    print("-" * (len(feature_type) + 28))
    print(results_df)
    
    # Plot significant correlations
    significant_features = results_df[results_df['p_value'] < 0.05].index
    for feature in significant_features:
        plt.figure(figsize=(12, 6))
        
        # Create scatter plot
        sns.scatterplot(data=analysis_df, x=feature, y='ladder_score')
        
        # Add regression line
        x = analysis_df[feature]
        y = analysis_df['ladder_score']
        z = np.polyfit(x, y, 1)
        p = np.poly1d(z)
        plt.plot(x, p(x), "r--", alpha=0.8)
        
        # Add title and labels
        plt.title(f'{feature} vs Happiness Score\n' + 
                 f'r={results_df.loc[feature, "correlation"]:.3f}, ' +
                 f'p={results_df.loc[feature, "p_value"]:.3f}')
        plt.xlabel(feature)
        plt.ylabel('Happiness Score')
        
        # Add country labels for potential outliers
        residuals = y - p(x)
        std_residuals = residuals / residuals.std()
        for idx, (std_resid, country) in enumerate(zip(std_residuals, analysis_df['country'])):
            if abs(std_resid) > 2:  # Label points > 2 std from regression
                plt.annotate(country, (x.iloc[idx], y.iloc[idx]))
        
        plt.tight_layout()
        plt.show()
```

```python
def analyze_musical_features_vs_happiness():
    """Main analysis function"""
    # Merge happiness data with musical features
    analysis_df = pd.merge(
        wh_data[['country', 'ladder_score'] + happiness_components],
        songs_data_filtered_stats_country,
        on='country'
    )
    
    # Analyze each feature type
    analyze_correlations(analysis_df, base_features, "Base Musical Features")
    analyze_correlations(analysis_df, std_features, "Musical Variability")
    analyze_correlations(analysis_df, diversity_features, "Musical Diversity")
    
    return analysis_df  # Return for additional analysis if needed
```

```python
analysis_df = analyze_musical_features_vs_happiness()
```

### Initial observations <a name="initial-observations"></a>

#### Base Musical Features Observations:
- Strongest negative correlatoin with danceability (-0.368, p<0.01): Happier countries tend to listen to less danceable music.
- Strong positive correlation with mode_pct (0.366, p<0.01): Happier countries tend to listen to more songs in major keys.
- Moderate correlations with tempo (positive), speechiness, energy (both negative)

#### Musical Variability Observations:
- Strong positive correlations across multiple features (energy, acousticness, liveness, danceability).
- This suggests happier countries tend to have more varied musical preferences.
- Particuarly strong for energy_std (0.359, p<0.01).

#### Musical Diversity Observations:
- Strongest correlations of all three categories.
- liveness_diversity (0.436, p<0.001) and speechiness_diversity (0.426, p<0.001) show the strongest relationships.
- Interestingly, acousticness_diversity shows a negative correlation (-0.238, p<0.05).

#### Key insights:
- Musical variety appears more strongly related to happiness than absolute musical characteristics.
- Happier countries tend to:
    - Listen to more songs in major keys
    - Have more diverse musical preferences
    - Show greater variation in their musical choices
    - Prefer less danceable music (which is somewhat counterintuitive).

## 4 - Distribution analysis <a name="distribution-analysis"></a>

In this section, we will examine the distribution of happiness scores, look at distributions of key musical features, and create box plots to compare these across happiness quartiles.

```python
def analyze_happiness_distribution(analysis_df):
    """Analyze the distribution of happiness scores"""
    plt.figure(figsize=(12, 6))
    
    # Create subplot with both histogram and kernel density
    sns.histplot(data=analysis_df, x='ladder_score', bins=15, kde=True)
    plt.title('Distribution of Happiness Scores')
    plt.xlabel('Happiness Score')
    
    # Add descriptive statistics as text
    stats_text = f'Mean: {analysis_df["ladder_score"].mean():.2f}\n'
    stats_text += f'Median: {analysis_df["ladder_score"].median():.2f}\n'
    stats_text += f'Std: {analysis_df["ladder_score"].std():.2f}\n'
    stats_text += f'Skew: {analysis_df["ladder_score"].skew():.2f}'
    
    plt.text(0.95, 0.95, stats_text,
             transform=plt.gca().transAxes,
             verticalalignment='top',
             horizontalalignment='right',
             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Save the figure to a file
    plt.savefig(f"./Figures/2.2 happiness distribution.png", dpi=300, bbox_inches='tight')
    
    plt.show()
    
    # Print top and bottom 5 countries by happiness
    print("\nTop 5 Happiest Countries:")
    print(analysis_df.nlargest(5, 'ladder_score')[['country', 'ladder_score']])
    
    print("\nBottom 5 Countries by Happiness:")
    print(analysis_df.nsmallest(5, 'ladder_score')[['country', 'ladder_score']])
```

```python
def analyze_feature_distributions(analysis_df):
    """Analyze distributions of individual musical features"""
    # Features to analyze (same as above)
    features_to_plot = [
        'danceability', 'mode_pct', 'tempo', 'speechiness', 'energy',
        'energy_std', 'acousticness_std', 'liveness_std',
        'liveness_diversity', 'speechiness_diversity'
    ]
    
    # Create subplots for each feature
    n_features = len(features_to_plot)
    n_cols = 2
    n_rows = (n_features + 1) // 2
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4*n_rows))
    axes = axes.flatten()
    
    for i, feature in enumerate(features_to_plot):
        sns.histplot(data=analysis_df, x=feature, kde=True, ax=axes[i])
        axes[i].set_title(f'Distribution of {feature}')
        
        # Add basic statistics
        stats_text = f'Mean: {analysis_df[feature].mean():.2f}\n'
        stats_text += f'Std: {analysis_df[feature].std():.2f}'
        axes[i].text(0.95, 0.95, stats_text,
                    transform=axes[i].transAxes,
                    verticalalignment='top',
                    horizontalalignment='right',
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Remove any empty subplots
    for i in range(len(features_to_plot), len(axes)):
        fig.delaxes(axes[i])
    
    # Save the figure to a file
    plt.savefig(f"./Figures/2.3 feature distributions.png", dpi=300, bbox_inches='tight')
    
    plt.tight_layout()
    plt.show()
```

```python
def analyze_musical_features_by_happiness_quartile(analysis_df):
    """Create violin plots for musical features across happiness quartiles"""
    # Create happiness quartiles
    analysis_df['happiness_quartile'] = pd.qcut(analysis_df['ladder_score'], 
                                              q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
    
    # Features to analyze (focusing on those with significant correlations)
    features_to_plot = [
        'danceability', 'mode_pct', 'tempo', 'speechiness', 'energy',
        'energy_std', 'acousticness_std', 'liveness_std',
        'liveness_diversity', 'speechiness_diversity'
    ]
    
    # Create violin plots
    for feature in features_to_plot:
        plt.figure(figsize=(12, 6))
        sns.violinplot(data=analysis_df, x='happiness_quartile', y=feature)
        plt.title(f'Distribution of {feature} by Happiness Quartile')
        
        # Add ANOVA test results
        f_stat, p_val = stats.f_oneway(*[group[feature].values 
                                        for name, group in analysis_df.groupby('happiness_quartile')])
        plt.text(0.95, 0.95, f'ANOVA:\nF-stat: {f_stat:.2f}\np-value: {p_val:.3f}',
                transform=plt.gca().transAxes,
                verticalalignment='top',
                horizontalalignment='right',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        # Save the figure to a file with the feature name in the filename
        sanitized_feature_name = feature.replace(" ", "_").replace("/", "_")
        plt.savefig(f"./Figures/{sanitized_feature_name}_by_happiness_quartile.png", dpi=300, bbox_inches='tight')
        
        plt.show()
```

## 5 - Happiness components analysis <a name="happiness-components-analysis"></a>

In this section, we will analyze how musical features correlate with each happiness components and see whether certain components have stronger relationships with certain musical characteristics. We will also identify any interesting patterns that emerge in our analysis.

Our analysis will:
- Create correlation matrices between musical features and each happiness component
- Generate heatmaps showing significant correlations
- Identify the strongest relationships
- Separate the analysis by base features, variability, and diversity

```python
def analyze_components_correlations(analysis_df, features, components):
    """Create correlation matrix between musical features and happiness components"""
    # Calculate correlations
    correlations = []
    p_values = []
    
    for feature in features:
        feature_correlations = []
        feature_p_values = []
        for component in components:
            corr, p_val = stats.pearsonr(analysis_df[feature], analysis_df[component])
            feature_correlations.append(corr)
            feature_p_values.append(p_val)
        correlations.append(feature_correlations)
        p_values.append(feature_p_values)
    
    # Create correlation DataFrame
    corr_df = pd.DataFrame(correlations, 
                          index=features, 
                          columns=components)
    
    # Create p-values DataFrame
    p_df = pd.DataFrame(p_values,
                       index=features,
                       columns=components)
    
    # Add significance indicators to correlation values
    corr_with_sig = corr_df.round(3).astype(str).copy()
    for i in range(len(features)):
        for j in range(len(components)):
            stars = '***' if p_values[i][j] < 0.001 else \
                   '**' if p_values[i][j] < 0.01 else \
                   '*' if p_values[i][j] < 0.05 else ''
            corr_with_sig.iloc[i,j] = f"{correlations[i][j]:.3f}{stars}"
    
    return corr_df, p_df, corr_with_sig
```

```python
def plot_component_heatmap(corr_df, p_df, title):
    """Create heatmap of correlations with significance indicators"""
    plt.figure(figsize=(15, 10))
    
    # Create mask for non-significant correlations
    mask = p_df >= 0.05
    
    # Create heatmap
    sns.heatmap(corr_df, 
                annot=True, 
                cmap='coolwarm', 
                center=0,
                fmt='.3f',
                mask=mask,
                vmin=-1, vmax=1)
    
    plt.title(title)
    plt.xticks(rotation=45, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    
    # Save the figure to a file
    plt.savefig(f"./Figures/2.5 component heatmap - {title}.png", dpi=300, bbox_inches='tight')
    
    plt.show()
```

```python
# Get happiness components (excluding ladder score and dystopia)
components = [col for col in happiness_components]
```

```python
# Analyze each type of musical feature
print("Correlations between Musical Features and Happiness Components")
print("-----------------------------------------------------------")

# Base features analysis
base_corr, base_p, base_sig = analyze_components_correlations(
    analysis_df, base_features, components)

print("\nBase Musical Features:")
print(base_sig)
plot_component_heatmap(base_corr, base_p, "Base Musical Features vs. Happiness Components")

# Variability analysis
var_corr, var_p, var_sig = analyze_components_correlations(
    analysis_df, std_features, components)

print("\nMusical Variability:")
print(var_sig)
plot_component_heatmap(var_corr, var_p, "Musical Variability vs. Happiness Components")

# Diversity analysis
div_corr, div_p, div_sig = analyze_components_correlations(
    analysis_df, diversity_features, components)

print("\nMusical Diversity:")
print(div_sig)
plot_component_heatmap(div_corr, div_p, "Musical Diversity vs. Happiness Components")
```

```python
# Identify strongest relationships
def print_top_correlations(corr_df, p_df, n=5):
    """Print top n strongest significant correlations"""
    # Flatten correlation and p-value matrices
    flat_corr = corr_df.unstack()
    flat_p = p_df.unstack()
    
    # Create DataFrame with feature, component, correlation, and p-value
    results = pd.DataFrame({
        'correlation': flat_corr,
        'p_value': flat_p
    })
    results.index.names = ['Feature', 'Component']
    results = results.reset_index()
    
    # Filter for significant correlations and sort by absolute correlation
    significant = results[results['p_value'] < 0.05]
    top = significant.nlargest(n, 'correlation')
    bottom = significant.nsmallest(n, 'correlation')
    
    print(f"\nTop {n} Positive Correlations:")
    for _, row in top.iterrows():
        print(f"{row['Feature']} vs {row['Component']}: r={row['correlation']:.3f}, p={row['p_value']:.3f}")
    
    print(f"\nTop {n} Negative Correlations:")
    for _, row in bottom.iterrows():
        print(f"{row['Feature']} vs {row['Component']}: r={row['correlation']:.3f}, p={row['p_value']:.3f}")
```

```python
print("Strongest Relationships:")
print_top_correlations(pd.concat([base_corr, var_corr, div_corr]), 
                      pd.concat([base_p, var_p, div_p]))
```

### Observations <a name="hca-observations"></a>
Most notable positive correlations:
- **Generosity and Major Keys (`mode_pct`)**: The strongest positive correlation (r=0.553) suggests that countries with higher generosity tend to listen to more songs in major keys.
- **GDP and Musical Variability:** Strong correlations with both `energy_std` (r=0.528) and `danceability_std` (r=0.460) indicate that wealthier countries have more varied musical preferences.
- **Generosity and Musical Variability:** Strong correlations with generosity also exist with both `acousticness_std` (r=0.519) and `energy_std` (r=0.454).

Most notable negative correlations:
- **GDP and Danceability:** The strongest negative correlation (r=-0.591) suggests that wealthier countries tend to listen to less danceable music.
- **Generosity and Musical Features:** Shows consistent negative correlations with:
    - Danceability (r=-0.474)
    - Valence (r=-0.446)
    - Speechiness (r=-0.432)

Both GDP and generosity appear to have the strongest relationships with musical preferences, but in different ways. While both correlate negatively with danceability, they also show strong positive correlations with musical variability, suggesting that more prosperous and generous societies tend to have more diverse musical tastes.
