## Part 1: Exploratory data analysis <a name="p1-eda"></a>

### Import relevant packages

```python
import pandas as pd
import numpy as np
import json
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read JSON file

```python
# Open and read the JSON file
with open('logins.json', 'r') as file:
    data = json.load(file)
```

```python
# Check keys of the dictionary
data.keys()
```

```python
# Check contents of 'login_time'
for i in data['login_time']:
    print(i)
```

```python
# Create DataFrame from data
logins_df = pd.DataFrame(data=data)
```

```python
# Check info of logins_df
logins_df.info()
```

```python
# Convert logins_df['login_time'] to datetime
logins_df['login_time'] = pd.to_datetime(logins_df['login_time'])
```

```python
# Check info again
logins_df.info()
```

### Aggregate .json file based on 15-minute time intervals

```python
# Add a column 'count' to logins_df so we can aggregate login counts
logins_df['count'] = 1
```

```python
df_15min = logins_df.groupby(pd.Grouper(key='login_time', freq='15min'))['count'].sum().reset_index()
```

### Visualize and describe resulting time series of login counts
Highlight ways that best characterize the underlying patterns of the demand.

```python
# Add additional time-based columns for analysis
df_15min['hour'] = df_15min['login_time'].dt.hour
df_15min['day_of_week'] = df_15min['login_time'].dt.day_name()
df_15min['date'] = df_15min['login_time'].dt.date
```

```python
# Calculate daily patterns
daily_pattern = df_15min.groupby('hour')['count'].mean()
```

```python
# Calculate day-of-week patterns
weekly_pattern = df_15min.groupby('day_of_week')['count'].mean()
```

```python
# Calculate basic statistics
total_logins = df_15min['count'].sum()
mean_logins = df_15min['count'].mean()
max_logins = df_15min['count'].max()
std_logins = df_15min['count'].std()
```

```python
# Check for missing intervals
full_range = pd.date_range(start=df_15min['login_time'].min(),
                         end=df_15min['login_time'].max(),
                         freq='15min')
missing_intervals = full_range.difference(df_15min['login_time'])
```

```python
# Set figure style and size
plt.style.use('ggplot')
fig = plt.figure(figsize=(20, 12))

# 1. Time series plot - showing first week only for better visibility
plt.subplot(2, 2, 1)
one_week_data = df_15min.iloc[:672]  # First week (24*4*7 = 672 15-minute intervals)
plt.plot(one_week_data['login_time'], one_week_data['count'], linewidth=1)
plt.title('Login Counts Over First Week (15-minute intervals)')
plt.xlabel('Time')
plt.ylabel('Number of Logins')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 2. Daily pattern
plt.subplot(2, 2, 2)
daily_pattern = df_15min.groupby('hour')['count'].mean()
daily_pattern.plot(kind='bar', color='skyblue')
plt.title('Average Login Counts by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Number of Logins')
plt.grid(True, alpha=0.3)

# 3. Weekly pattern
plt.subplot(2, 2, 3)
weekly_pattern = df_15min.groupby('day_of_week')['count'].mean()
# Reorder days to start with Monday
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_pattern = weekly_pattern.reindex(days_order)
weekly_pattern.plot(kind='bar', color='lightgreen')
plt.title('Average Login Counts by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Average Number of Logins')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 4. Distribution of login counts
plt.subplot(2, 2, 4)
sns.histplot(data=df_15min, x='count', bins=30, color='salmon')
plt.title('Distribution of Login Counts per 15-min Interval')
plt.xlabel('Number of Logins')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

**Observations:**
- In the first week, login counts generally range from 0 to around 20, peaking at over 30 on 1970-01-04.
- Looking at average login counts by day of the week, the number of logins increases throughout the week, starting at around 6 logins on Monday and peaking at around 13 logins on Saturday. Sundays average just over 12 logins.
- Looking at average login counts by hour of the day, the average number of logins is lowest at 7 and 8 am (around 2 logins). Average logins climb to around 14 at 11am before gradually decreasing to around 6 logins at 5pm. From there, they gradually climb to a maximum of 16 logins at 10pm.
- Our distribution of login counts per 15-minute interval suggests that there are generally fewer than 10 logins per 15-minute interval. It is rare to see over 30 logins in a 15-minute interval.

```python
# Print summary statistics
print("Summary Statistics:")
print(f"Average logins per 15 minutes: {df_15min['count'].mean():.2f}")
print(f"Maximum logins in any 15-minute period: {df_15min['count'].max()}")
print(f"Standard deviation: {df_15min['count'].std():.2f}")
print(f"Median logins per 15 minutes: {df_15min['count'].median()}")

# Calculate and print peak hours
peak_hours = daily_pattern.nlargest(3)
print("\nPeak Hours:")
for hour, count in peak_hours.items():
    print(f"{hour:02d}:00 - Average of {count:.1f} logins per 15-min interval")

# Calculate and print busiest days
busiest_days = weekly_pattern.nlargest(3)
print("\nBusiest Days:")
for day, count in busiest_days.items():
    print(f"{day}: Average of {count:.1f} logins per 15-min interval")
```

These summary statistics reinforce what we observed in our visualizations. 
- Generally, there are fewer than 10 logins per 15-minutes; however, the high standard deviation of 8.33 relative to the mean of 9.52 indicates high variability.
- Peak hours are 10pm at 16.2 average logins per interval, 1pm at 15.5 average logins, and 11pm at 14.8 average logins.
- Busiest days are weekends: Saturday at 13.5 average logins per interval, Sunday at 12.6 average logins, and Friday at 10.6 average logins.

#### Weekly patterns

```python
# Analyze weekly patterns
# Reorder days to start with Monday
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

daily_stats = df_15min.groupby('day_of_week')['count'].agg(['mean', 'median', 'std', 'count']).round(2)
daily_stats = daily_stats.reindex(days_order)

print("Weekly Pattern Analysis:")
print("\nDaily Statistics:")
print(daily_stats)
```

## Part 2: Experiment and metrics design <a name="p2-design"></a>

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.

However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.

<div class="alert alert-block alert-info">
    <b>Question 1:</b>
    What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
    </div>

The key metric of success that I would use in this experiment, assuming we have access to this information, is the percentage of peak-time demand met in each city. Having drivers available from each city would help meet demand spikes. If a significantly higher percentage of demand is being met after implementing the plan than before, then we will know that reimbursing toll costs for drivers is a reasonable solution for our problem. 

<div class="alert alert-block alert-info">
    <b>Question 2:</b> Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:<br>
    &nbsp;&nbsp;&nbsp;&nbsp;a. how you will implement the experiment<br>
    &nbsp;&nbsp;&nbsp;&nbsp;b. what statistical test(s) you will conduct to verify the significance of the observation<br>
    &nbsp;&nbsp;&nbsp;&nbsp;c. how you would interpret the results and provide recommendations to the city operations team along with any caveats.
</div>

<div class="alert alert-block alert-info">
<b>2a: Provide details on how you will implement the experiment.</b>
</div>

To implement the experiment, we will first need to consider the following:
- Since we are collecting data on meeting peak-time demand, we will need to clearly define when "peak time" is. In order to do this, we will need to collect several weeks of baseline data for both cities and carefully examine how demand for driver services varies 1. according to the city, and 2. according to the time of day.
- We would also need to consider the timing of our experiment and carefully choose when to implement it. Special events or variations on seasonal demand could skew our data.
- We will also need to implement a system to track and process toll reimbursements.

Furthermore, rather than rolling out toll reimbursement all at once, we will split our drivers into control and treatment groups. Our treatment group will receive toll reimbursement, while our control group will not.

With that being said, we could approach this experiment as follows:
- Collect several weeks of baseline data on peak-time demand fulfillment.
- Randomly select a subset of drivers to receive toll reimbursement, and keep other drivers as a control group.
- Run the experiment for several weeks (as many weeks worth of baseline data) to gather comparison data.
- Examine what percentage of peak demand each group is able to meet.

<div class="alert alert-block alert-info">
<b>2b: What statistical test(s) will you conduct to verify the significance of the observation?</b>
</div>

The statistical test we run will ultimately depend on the distribution of the data. Assuming a normal distribution, we can run a two-sample t-test, which would compare the mean percentage of demand met between the two groups.

However, if our data is not normally distributed, we could use a chi-square test to compare the porportion of fulfilled vs. unfulfilled demand between groups.

Also, considering activity varies between weekdays and weekends, we might conduct statistical tests separately on weekend vs. weekday data. Likewise, we might conduct statistical tests separately on Gotham vs. Metropolis data.

<div class="alert alert-block alert-info">
<b>2c: How would you interpret the results and provide recommendations to the city operations team (along with any caveats)?</b>
</div>

We would recommend **full implementation** of the plan if:
- Peak-time demand fulfillment shows statistically significant improvement,
- The improvement is substantial enough to justify the cost, and
- Benefits are seen in both cities.
With this recommendation, we would also propose a gradual adoption strategy in order to manage costs and monitor long-term effects of the program. Furthermore, we would create clear reimbursement guidelines for the city operations team to follow.
<br><br>

However, if results are mixed, we might consider **targeted implementation**, in which we would only reimburse tolls during specific times (such as weekends or transition periods) or focus on areas where data shows the most benefit.<br><br>

If results show **no improvement**, we would recommend an **alternative approach**. Perhaps the current city-specific specialization is the optimal arrangement, or we could explore other ways to meet demand spikes.

**Caveats and points to consider:**
- Financial considerations: Does the cost of the toll reimbursement justify the benefit of improved demand fulfillment?<br><br>

- We could encounter a spillover effect issue if drivers in the control group communicate with drivers in the treatment group. For example, control group drivers who do not receive toll reimbursement could go on strike out of perceived unfair treatment.
    - We will need to communicate, therefore, that if successful, this experiment will ultimately benefit all drivers.
    - We might also consider keeping the experimental period short to minimize frustration.<br><br>
    
    
- On weekdays, since Gotham is active at night and Metropolis is active during the day, having drivers stick to one city might be the most logical and efficient approach. We need to consider whether the core assumption that drivers should serve both cities is even correct and whether the toll reimbursement program is addressing a real problem.
    - However, we also need to consider:
        - Weekends, when both cities have reasonable activity.
        - Transition periods (for example, late afternoon or early evening) when activity might overlap.
        - The drivers' personal preferences for when they work (for example, drivers from Metropolis who prefer the night shift).
