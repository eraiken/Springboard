```python
def analyze_split_datasets(final_df, elem_middle_schools, high_schools):
    """
    Split and analyze data quality separately for elementary/middle schools and high schools
    """
    # Split the final dataset
    elem_middle_data = final_df[final_df['DBN'].isin(elem_middle_schools['DBN'])]
    high_school_data = final_df[final_df['DBN'].isin(high_schools['DBN'])]
    
    # Define relevant metrics for each school type
    elem_middle_metrics = [
        # Achievement metrics
        'ela_proficient_pct',
        'math_proficient_pct',
        'ela_avg_proficiency',
        'math_avg_proficiency',
        'ela_lowest_third_proficiency',
        'math_lowest_third_proficiency',
        
        # Demographic metrics
        'enrollment',
        'ell_pct',
        'disability_pct',
        'asian_pct',
        'black_pct',
        'hispanic_pct',
        'white_pct',
        
        # Attendance metrics
        'student_attendance_rate',
        'chronic_absence_pct',
        'teacher_attendance_rate'
    ]

    high_school_metrics = [
        # Achievement metrics
        'grad_rate_4yr',
        'grad_rate_6yr',
        'college_ready_4yr',
        'postsec_enroll_6mo',
        'regents_english',
        'regents_algebra',
        'regents_living_env',
        
        # Demographic metrics
        'enrollment',
        'ell_pct',
        'disability_pct',
        'asian_pct',
        'black_pct',
        'hispanic_pct',
        'white_pct',
        
        # Attendance metrics
        'student_attendance_rate',
        'chronic_absence_pct',
        'teacher_attendance_rate'
    ]
    
    print("="*80)
    print(f"Data Quality Analysis Summary")
    print("="*80)
    
    print(f"\nTotal schools in dataset: {len(final_df['DBN'].unique())}")
    print(f"Elementary/Middle schools: {len(elem_middle_data['DBN'].unique())}")
    print(f"High schools: {len(high_school_data['DBN'].unique())}")
    
    # Analyze elementary/middle schools
    print("\n" + "="*30 + " Elementary/Middle Schools " + "="*30)
    missing_elem = analyze_data_quality(
        elem_middle_data[elem_middle_metrics], 
        "Elementary/Middle Schools"
    )
    
    # Analyze high schools
    print("\n" + "="*30 + " High Schools " + "="*30)
    missing_hs = analyze_data_quality(
        high_school_data[high_school_metrics],
        "High Schools"
    )
    
    return elem_middle_data, high_school_data, missing_elem, missing_hs

def create_split_visualizations(elem_middle_data, high_school_data):
    """
    Create separate visualizations for elementary/middle schools and high schools
    """
    fig = plt.figure(figsize=(20, 15))
    
    # 1. Academic Year Distribution
    plt.subplot(3, 2, 1)
    elem_middle_data['academic_year'].value_counts().sort_index().plot(kind='bar')
    plt.title('Academic Years - Elementary/Middle Schools')
    plt.xticks(rotation=45)
    
    plt.subplot(3, 2, 2)
    high_school_data['academic_year'].value_counts().sort_index().plot(kind='bar')
    plt.title('Academic Years - High Schools')
    plt.xticks(rotation=45)
    
    # 2. Attendance Rate Distributions
    plt.subplot(3, 2, 3)
    sns.boxplot(data=elem_middle_data, y='student_attendance_rate')
    plt.title('Student Attendance - Elementary/Middle Schools')
    
    plt.subplot(3, 2, 4)
    sns.boxplot(data=high_school_data, y='student_attendance_rate')
    plt.title('Student Attendance - High Schools')
    
    # 3. Achievement Score Distributions
    plt.subplot(3, 2, 5)
    if 'ela_proficient_pct' in elem_middle_data.columns:
        sns.boxplot(data=elem_middle_data[['ela_proficient_pct', 'math_proficient_pct']])
        plt.title('ELA/Math Proficiency - Elementary/Middle Schools')
    
    plt.subplot(3, 2, 6)
    if 'grad_rate_4yr' in high_school_data.columns:
        sns.boxplot(data=high_school_data[['grad_rate_4yr', 'college_ready_4yr']])
        plt.title('Graduation/College Readiness - High Schools')
    
    plt.tight_layout()
    plt.show()

# Run the analysis
try:
    # Split and analyze the data
    elem_middle_data, high_school_data, missing_elem, missing_hs = analyze_split_datasets(
        final_df, elem_middle_schools, high_schools
    )
    
    # Create visualizations
    create_split_visualizations(elem_middle_data, high_school_data)
    
except Exception as e:
    print(f"An error occurred: {str(e)}")
    print("\nLet's examine the data structure:")
    print("\nColumns in final_df:", final_df.columns.tolist())
```
